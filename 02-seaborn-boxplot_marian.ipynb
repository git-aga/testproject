```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("http://51.91.120.89/extras/pokemon.csv")
df
```

```python
df.groupby("Type 1")["Attack"].agg([pd.np.mean, pd.np.std])
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df, x="Attack", y="Type 1", orient="h", color="red", whis=1)
```

```python
plt.figure(figsize=(20,7))
sns.boxplot(data=df, y="Attack", x="Type 1", orient="v", whis=1, hue="Legendary")
```
