```python
#creatign pandas dataframe
results = [] 

for row in rows:
    rank = row.findAll("span")[0]["data-value"]
    title  = row.find("td", class_ ="titleColumn").find("a").text
    year = row.find("td", class_ ="titleColumn").find("span").text.strip("()")
    rating = row.find("td", class_ ="titleColumn").find("span").text.strip("()")
    url = "https://www.imdb.com/" + row.find("td", class_ ="titleColumn").find("a")["href"]
    print("rank", rank)
    print("title", title)
    print("year", year)
    print("rating",rating )
    print("url:",url)
    print("*"*60)
    
    
    movie = {
        "rank" : rank, 
        "title" : title, 
        "year" : year,
        "rating" : rating,
        "url" : url
    }
    results.append(movie)
    
```
