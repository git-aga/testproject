```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("http://51.91.120.89/extras/pokemon.csv")
df.sample(n=10)
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(df.Attack, color="r", shade=True, bw=10)
sns.kdeplot(df.Defense, color="g", shade=True, bw=10)
plt.show()
```

```python
plt.subplots(3, 2, figsize=(15,10))
for index, _bw in enumerate([0.1, 1, 10, 50, 200, 500], 1):
    plt.subplot(3, 2, index)
    sns.kdeplot(df.Attack, bw=_bw, color='red', shade=True).set_title(f"BW={_bw}")
```
